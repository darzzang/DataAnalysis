##### 다른방법---

```python
df_item_grp = df.groupby('item_name')[['quantity', 'item_price']].sum()
df_item_grp['item_unit_price'] = df_item_grp['item_price'] / df_item_grp['quantity']
df_item_grp
```

- 주문 수량이 1인 아이템은 item_price 가 동일한가

```python
df_quan_1  = df[df.quantity == 1]
df_quan_1_grp = df_quan_1.groupby('item_name')['item_price'].agg(['max','min'])
df_quan_1_grp
```

```python
df_quan_1_grp[df_quan_1_grp['max'] != df_quan_1_grp['min']]
```

```python
df_quan_1_grp[df_quan_1_grp['max'] != df_quan_1_grp['min']].shape
```

```python
df_quan_1_grp.colums = ['df_quan_1_grp_max','df_quan_1_grp_min']
df_quan_1_grp
```

```python
# 두 데이터 merge
# pd.merge(df_item_grp, df_quan_1_grp, on='item_name', how='inner')
df_quan_1_grp_mg = pd.merge(df_item_grp, df_quan_1_grp, on='item_name', how='outer')
df_quan_1_grp_mg
```

```python
# 시각화
plt.figure(figsize=(12,5))
plt.bar(df_quan_1_grp_mg.index, df_quan_1_grp_mg['min'].values)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df_quan_1_grp_mg.index, df_quan_1_grp_mg['item_unit_price'].values, label='mean')
plt.plot(df_quan_1_grp_mg.index, df_quan_1_grp_mg['max'].values, label='max')
plt.plot(df_quan_1_grp_mg.index, df_quan_1_grp_mg['min'].values, label='min')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.hist(df_quan_1_grp_mg.index, df_quan_1_grp_mg['min'].values, bins=20)
plt.xticks(rotation=90)
plt.show()
```
